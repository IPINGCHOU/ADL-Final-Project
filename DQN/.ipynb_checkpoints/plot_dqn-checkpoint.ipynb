```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##for dqn

mov_avg_dqn = []
mov_avg_dqn2 = []
mov_avg_dqn3 = []
mov_avg_dqn4 = []
mov_avg_doubledqn = []
mov_avg_duelingdqn = []

with open("./mov_avgs/mov_avg_dqn.txt", "r") as f:
    for line in f:
        mov_avg_dqn.append(float(line.strip()))
'''
with open("./mov_avgs/mov_avg_dqn2.txt", "r") as f:
    for line in f:
        mov_avg_dqn2.append(float(line.strip()))

with open("./mov_avgs/mov_avg_dqn3.txt", "r") as f:
    for line in f:
        mov_avg_dqn3.append(float(line.strip()))
        
with open("./mov_avgs/mov_avg_dqn4.txt", "r") as f:
    for line in f:
        mov_avg_dqn4.append(float(line.strip()))
'''

with open("./mov_avgs/mov_avg_doubledqn2.txt", "r") as f:
    for line in f:
        mov_avg_doubledqn.append(float(line.strip()))
        
with open("./mov_avgs/mov_avg_duelingdqn2.txt", "r") as f:
    for line in f:
        mov_avg_duelingdqn.append(float(line.strip()))


episodes = [x+1+(100-1) for x in list(range(len(mov_avg_dqn)))]
episodes2 = [x+1+(100-1) for x in list(range(len(mov_avg_dqn2)))]
episodes3 = [x+1+(100-1) for x in list(range(len(mov_avg_dqn3)))]
episodes4 = [x+1+(100-1) for x in list(range(len(mov_avg_dqn4)))]

episodes_double = [x+1+(100-1) for x in list(range(len(mov_avg_doubledqn)))]
episodes_dueling = [x+1+(100-1) for x in list(range(len(mov_avg_duelingdqn)))]

plt.plot(episodes[::], mov_avg_dqn[::],',-',color = 'r', label="dqn",linewidth = 1, alpha=1)
#plt.plot(episodes2[::], mov_avg_dqn2[::],',-',color = 'g', label="0.1",linewidth = 0.5, alpha=1)
#plt.plot(episodes3[::], mov_avg_dqn3[::],',-',color = 'b', label="0.05",linewidth = 0.5, alpha=0.8)
#plt.plot(episodes4[::], mov_avg_dqn4[::],',-',color = 'y', label="0.001",linewidth = 0.5, alpha=0.8)

#plt.plot(episodes_double[::], mov_avg_doubledqn[::],',-',color = 'b', label="double dqn",linewidth = 1)
#plt.plot(episodes_dueling[::], mov_avg_duelingdqn[::],',-',color = 'g', label="dueling dqn",linewidth = 1)

plt.title("Learning Curve of Rewards on MsPacman")
plt.xlabel("Episodes")
plt.ylabel("100 Episodes Moving Average Rewards")
plt.xticks(np.arange(0, 14000, 2000))
plt.yticks(np.arange(0, 60, 10))
plt.legend(loc = "best", fontsize=10)
plt.savefig('learning_curve_dqn.jpg', dpi=120)
plt.show
```

```python
plt.clf()

#for pg
mov_avg_pg = []

with open("./mov_avgs/mov_avg_pg.txt", "r") as f:
    for line in f:
        mov_avg_pg.append(float(line.strip()))

episodes = list(range(len(mov_avg_pg)))
episodes = [x+1+(100-1) for x in episodes]        

plt.plot(episodes, mov_avg_pg,',-',color = 'r', label="pg")

plt.title("Learning Curve of Rewards (100 episode window)")
plt.xlabel("Episodes")
plt.ylabel("Average Rewards")
plt.xticks(np.arange(0, 14000, 2000))
plt.yticks(np.arange(-200, 400, 50))
plt.legend(loc = "best", fontsize=10)
plt.savefig('learning_curve_pg.jpg', dpi=120)
plt.show
```
