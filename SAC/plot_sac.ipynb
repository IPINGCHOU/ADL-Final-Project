```python
def mov_average(a, n=100):
    window = n
    cumsum = [0]
    mov_avg = []


    for i, x in enumerate(a, 1):
        cumsum.append(cumsum[i-1] + x)
        if i >= window:
            avg = (cumsum[i] - cumsum[i-window])/window
            mov_avg.append(avg)
            
    return mov_avg
```
