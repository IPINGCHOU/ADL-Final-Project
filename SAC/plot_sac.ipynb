```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.load("sac_reward.npy")
print(len(a))

mov_avg_reward = mov_average(a)

episodes = list(range(len(mov_avg_reward)))
episodes = [x+1+(100-1) for x in episodes]   

plt.plot(episodes, mov_avg_reward,',-', label="sac")

plt.title("Learning Curve of Rewards (100 episode window)")
plt.xlabel("Episodes")
plt.ylabel("Average Rewards")
plt.xticks(np.arange(0, 1400, 200))
#plt.yticks(np.arange(-200, 400, 50))
plt.legend(loc = "best", fontsize=10)
#plt.savefig('./figure/learning_curve_sac.jpg', dpi=120)
plt.show
```

```python
b = np.load("sac_loss.npy", allow_pickle=True)
print(len(b))

mov_avg_loss = mov_average(b)

episodes = list(range(len(mov_avg_loss)))
episodes = [x+1+(100-1) for x in episodes]   

plt.plot(episodes, mov_avg_loss,',-', label="sac")

plt.title("Learning Curve of Loss (100 episode window)")
plt.xlabel("Episodes")
plt.ylabel("Average Loss")
plt.xticks(np.arange(0, 1400, 200))
#plt.yticks(np.arange(-200, 400, 50))
plt.legend(loc = "best", fontsize=10)
#plt.savefig('./figure/loss_sac.jpg', dpi=120)
plt.show
```

```python
def mov_average(a, n=100):
    window = n
    cumsum = [0]
    mov_avg = []


    for i, x in enumerate(a, 1):
        cumsum.append(cumsum[i-1] + x)
        if i >= window:
            avg = (cumsum[i] - cumsum[i-window])/window
            mov_avg.append(avg)
            
    return mov_avg
```
