```python
import matplotlib.pyplot as plt
import numpy as np
def moving_average(array, n) :
    ret = np.cumsum(array, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n 
```

```python
r=np.load('rw.npy')
plt.plot(moving_average(r, 100))
plt.title('MA of rewards n=100')
plt.ylabel('reward')
plt.xlabel('episode')
plt.show()
```

```python
loss=np.load('loss.npy')
plt.title('MA of loss n=100')
plt.ylabel('loss')
plt.xlabel('episode')
plt.plot(moving_average(loss, 100))
plt.show()
```
