```python
import matplotlib.pyplot as plt
import numpy as np
def moving_average(array, n) :
    ret = np.cumsum(array, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n 
```

```python
r=np.load('rw.npy')
print(len(r))
plt.plot(moving_average(r, 300))
```

```python
loss=np.load('loss.npy')
print(len(loss))
plt.plot(moving_average(loss, 100))
```
